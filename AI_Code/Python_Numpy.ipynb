1. Khởi tạo mảng (Array Creation)

```python
import numpy as np

# Tạo mảng từ một List
arr = np.array([1, 2, 3, 4, 5])

# Tạo mảng toàn số 0 (zeros) hoặc số 1 (ones)
zeros = np.zeros((2, 3)) # Mảng 2 hàng, 3 cột toàn số 0
ones = np.ones((3, 2)) # Mảng 3 hàng, 2 cột toàn số 1

# Tạo mảng với các giá trị cách đều nhau
# np.arange(start, stop, step)
range_arr = np.arange(0, 10, 2) # Kết quả: [0, 2, 4, 6, 8]

# Tạo mảng với số lượng phần tử xác định trong khoảng 
# np.linspace(start, stop, num)
linespace_arr = np.linspace(0, 1, 5) # 5 số cách đều nhau từ 0 đến 1

# Tạo mảng đơn vị (Identity matrix)
identity = np.eye(3) # Ma trận đơn vị cấp 3
```

2. Kiểm tra thuộc tính của mảng (Attributes)

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.shape) # (2, 3) - Số hàng và số cột
print(arr.ndim) # 2 - Số chiều của mảng
print(arr.size) # 6 - Tổng số phần tử
print(arr.dtype) # int64 - Kiểu dữ liệu của phần tử
```
