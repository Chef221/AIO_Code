Một số hàm thông dụng trong Streamlit

```python
import streamlit as st

# Tạo tiêu đề chính cho ứng dụng
st.title()

# Tạo công cụ tải tệp lên ứng dụng. Tham số type cho phép giới hạn định dạng tệp 
# được chấp nhận, ví dụ ["csv", "txt", "pdf"]
st.file_uploader()

# Dùng để hiển thị văn bản, bảng dữ liệu, biểu đồ hoặc các loại dữ liệu khác
st.write()

# Tạo thanh điều hướng bên trái, giúp tổ chức các widget và điều khiển hợp lý hơn
st.sidebar()

# Tạo nút bấm cho phép thực thi hành động khi người dùng nhắn vào 
st.button()

# Tạo biểu đồ tương tác sử dụng thư viện Plotly, hỗ trợ zoom, pan,...
st.plotly_chart()

# Các hàm tích hợp sẵn để trực quan hóa dữ liệu dưới dạng biểu đồ 
st.line_chart()
st.bar_chart()
st.area_chart()

# Hiển thị dữ liệu địa lý trên bản đồ tương tác, áp dụng cho các tập dữ liệu có tọa
# độ kinh độ và vĩ độ
st.map()
```
