```python
'''
- Khi tiếp cận một bài toán AI nói chung hay bài toán Object
Detection nói riêng thì việc tìm hiểu và nắm bắt được dữ liệu
là một điều quan trọng để có được chiến lược tiếp cận bài toán
và xây dựng model một cách hiệu quả

1. Khảo sát thông tin của ảnh trong bộ dữ liệu
- Một trong những thông tin cơ bản và cần ưu tiên nắm bắt đó
là kích thước của ảnh. Chúng ta cần biết được toàn bộ ảnh trong
dataset có cùng chung kích thước hay không

2. Thống kê số lượng bounding box (đối tượng cần phát hiện)
- Bounding box là một khái niệm cơ bản để đánh nhãn dữ liệu cho bài toán Object Detection
Ví dụ: Tưởng tượng có một bức ảnh chụp con mèo. Để máy tính tìm hiểu 'con mèo nằm ở đâu',
bạn sẽ vẽ một hình chữ nhật bao sát lấy con mèo đó => bouding box
- Máy tính không nhìn thấy hình chữ nhật, nó chỉ hiểu các con số. Trong dataset, một bouding
box thường được đại diện bằng một danh sách 4 số (tọa độ)
- Có 3 cách phổ biến để biểu diễn 4 số này:
+ Tọa độ góc (Pascal VOC format): [x_min, y_min, x_max, y_max]
+ Tọa độ + kích thước (COCO format): [x_min, y_min, width, height]
+ Tọa độ tâm chuẩn hóa (YOLO format): Tọa độ được chia cho kích thước ảnh để đưa về (0, 1)
[x_center, y_center, width, height]

=> Dữ liệu crowded: trong các bức ảnh có mật độ vật thể rất cao và chúng nằm sát lẫn nhau. 
(Ảnh chụp đám đông, đàn cừu)
- Crowded làm khó mô hình AI
+ Lỗi ở thuật toán NMS (Non-Maximum Suppression): Đây là kỹ thuật lọc bỏ các hộp trùng lặp.
Nếu AI thấy 2 hộp chồng lấn lên nhau quá nheieuf (IoU > 0.5) -> xóa bớt 1 hộp đi. Nhưng
khi crowded, có thể xóa nhầm người đứng sau => Miss detection
+ Nhiễu loạn đặc trưng (Feature Confusion)
+ Khó gán nhãn (Labelling Noise)

3. Thống kê về kích thước của bounding box
- Thông tin về kích thước của bounding box trong dataset sẽ giúp chúng ta thống kê được
phân phối về kích thước của đối tượng trong ảnh. Với những số liệu này chúng ta có thể đưa
ra kết luận được việc xử lý 'small object' hay không
- Đối với việc thống kê thì chúng ta cần thể hiện số liệu dưới dạng chart để có được cái nhìn
tổng quan hơn

4. Thống kê số lượng Object trong từng class
- Việc xử lý dữ liệu mất cân bằng không chỉ là vấn đề của bài toán Classification mà cũng
là vấn đề của bài toán Detection
'''
```

```python
'''
1. Sàng lọc cấp độ ảnh (Image Level Cleaning)
- Mục tiêu: Loại bỏ các tấm ảnh bị lỗi, kích thước quá khổ (outliers) có thể làm sai lệch
tỉ lệ khi resize
- Hành động: Vẽ biểu đồ -> Khoanh vùng an toàn -> Xóa bỏ các ảnh ngoại lai

2. Sàng lọc cấp độ vật thể (Object Level Sampling)
- Giả sử các vật thể quan trọng (ví dụ: xe_bus) chỉ xuất hiện trong các bức ảnh có kích thước 
cực lớn mà bạn vừa lỡ tay xóa đi ở Bước 1. Lúc này biểu đồ phân phối object ở Bước 2 sẽ bị 
mất hẳn một mảng (mất phân phối ban đầu).

Giải pháp A: Resize (Thu nhỏ ảnh)
Thay vì xóa bức ảnh 4000x3000, bạn resize nó về kích thước an toàn (ví dụ: 1024x768) và 
đồng thời tính lại tọa độ bounding box theo tỷ lệ thu nhỏ đó.

Ưu điểm: Giữ được dữ liệu.

Nhược điểm: Nếu vật thể trong ảnh vốn đã bé, khi thu nhỏ ảnh lại, vật thể có thể biến 
mất hoặc trở nên quá mờ (nhỏ hơn 32px) khiến model không học được.

Giải pháp B: Tiling / Sliding Window (Cắt nhỏ ảnh) - Khuyên dùng cho ảnh vệ tinh/drone
Thay vì xóa bức ảnh to, bạn cắt nó ra thành nhiều bức ảnh nhỏ (patches/tiles) có kích thước 
chuẩn (ví dụ: cắt ảnh 4000x3000 thành 12 bức 1000x1000).

Cách xử lý box: Giữ lại các box nằm trọn trong tấm ảnh con. Xử lý khéo léo các 
box nằm ở đường cắt (bị cắt đôi).

Ưu điểm: Không làm biến dạng vật thể, giữ nguyên độ phân giải nét.

Giải pháp C: Padding (Thêm viền)
Nếu ảnh có tỷ lệ quá dị (ví dụ ảnh panorama dài ngoằng: 2000x200), thay vì xóa, 
bạn thêm viền đen (padding) để nó trở thành hình vuông hoặc tỷ lệ chuẩn hơn trước 
khi đưa vào model.
'''
```
