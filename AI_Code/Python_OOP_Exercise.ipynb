```python
'''
Một Ward gồm có name (string) và danh sách của mọi người trong Ward. Một người person có thể 
là student, doctor, teacher. Một student gồm có name, yob (int) và grade (string). Một
teacher gồm có name, yob và subject (string). Một doctor gồm có name, yob, và specialist(string)
Lưu ý cần sử dụng a list để chứa danh sách của mọi người trong Ward

- (a) Thực hiện các class student, doctor, và teacher theo mô tả trên. Thực hiện describe()
để print ra tất cả các thông tin của các objects
- (b) Viết addPerson(person) method trong Ward class để add thêm một người mới với nghề 
nghiệp bất kỳ vào danh sách người của ward. Tạo ra một ward object, và thêm vào 1 student,
2 teacher và 2 doctor. Thực hiện describe() để in ra tên ward (name) và toàn bộ thông tin
của từng người trong ward
- (c) Viết countDoctor() method để đếm số lượng doctor trong ward
- (d) Viết sortAge() method để sort mọi người trong ward theo tuổi của họ với thứ tự 
tăng dần (có thể sử dụng sort của list hoặc viết thêm function đều được)
- (e) Viets aveTeacherYearOfBirth() method để tính trung bình năm sinh của các teachers trong
ward
'''

class Person:
    def __init__(self, name, yob):
        self._name = name
        self._yob = yob
    def describe(self):
        print(f'Name: {self._name} | Year of birth: {self._yob} |', end = ' ')
    def getJob(self):
        pass
    def getYob(self):
        return self._yob
class Student(Person):
    def __init__(self, name, yob, grade):
        super().__init__(name, yob)
        self.__grade = grade
    def describe(self):
        super().describe()
        print(f'Grade: {self.__grade}')
    def getJob(self):
        return 'Student'
class Teacher(Person):
    def __init__(self, name, yob, subject):
        super().__init__(name, yob)
        self.__subject = subject
    def describe(self):
        super().describe()
        print(f'Subject: {self.__subject}')
    def getJob(self):
        return 'Teacher'
class Doctor(Person):
    def __init__(self, name, yob, specialist):
        super().__init__(name, yob)
        self.__specialist = specialist
    def describe(self):
        super().describe()
        print(f'Specialist: {self.__specialist}')
    def getJob(self):
        return 'Doctor'
class Ward:
    def __init__(self, name):
        self.__name = name
        self.__listofPeople = []
    def addPerson(self, person):
        self.__listofPeople.append(person)
    def describe(self):
        print(f'Ward name: {self.__name}')
        for idx, person in enumerate(self.__listofPeople):
            print(f'Number {idx + 1}: ', end = ' ')
            person.describe()
    def countDoctor(self):
        doctorCount = 0
        for person in self.__listofPeople:
            if person.getJob() == 'Doctor':
                doctorCount += 1
        return doctorCount
    def sortAge(self):
        self.__listofPeople.sort(key = lambda x: x.getYob(), reverse = True)
    def aveTeacherYearOfBirth(self):
        sumOfYearOfBirth = 0
        countTeacher = 0
        for person in self.__listofPeople:
            if person.getJob() == 'Teacher':
                sumOfYearOfBirth += person.getYob()
                countTeacher += 1
        return sumOfYearOfBirth // countTeacher 

# (b)
print('\nQuestion B:--------------------------------------------')
newWard1 = Ward('Thanh Hoa')
studentOne = Student('Nguyen', 2006, 10)
teacherOne = Teacher('Thuong', 1980, 'IT003')
teacherTwo = Teacher('Hang', 1975, 'CS115')
doctorOne = Doctor('Hai', 1965, 'Brain')
doctorTwo = Doctor('Phuong', 1970, 'Cancer')
newWard1.addPerson(studentOne)
newWard1.addPerson(teacherOne)
newWard1.addPerson(teacherTwo)
newWard1.addPerson(doctorOne)
newWard1.addPerson(doctorTwo)
newWard1.describe()

# (c)
print('\nQuestion C:--------------------------------------------')
print(f'Number of doctor in the ward: {newWard1.countDoctor()}')

# (d)
print('\nQuestion D:--------------------------------------------')
newWard1.sortAge()
newWard1.describe()

# (e)
print('\nQuestion E:--------------------------------------------')
print(f'Average teacher year of birth: {newWard1.aveTeacherYearOfBirth()}')

        
```
