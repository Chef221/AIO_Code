```python
'''
Searching problem
- Input: a sorted sequence of n integer number <a1, a2,..., an> and key
- Output: index of key in the sequence if exist, -1 if not exist
'''
def Linear_Search(arr, key):
    n = len(arr)
    for idx in range(n):
        element = arr[idx]
        if element == key:
            return idx
    return -1
'''
Program = Data Structure + Algorithm
- Data structure: A way of organizing, stoning and performing operations on data (list,...)
- Algorithm: Describes a sequence of steps to solve a perform a calculation. Described
in pseudocode, a programming language
'''

# Algorithm
'''
Design an algorithm:
- Prove the algorithm is correct: Loop invariant, recursive function,...
Analysis the algorithm:
- Compare the efficiency of the algorithm
Sequential and parallel algorithms:
- Random access model (RAM)
- Parallel multi-processor access model (PRAM)

'''
```
