Exponential Functions

```python
'''
# Definition: 
- "Exponential functions model a relationship in which a constant change in the independent
variable gives the same proportional change in the dependent variable (Ex: y = e ** x)
- An exponential function can describe growth or decay
'''
import matplotlib.pyplot as plt
def func1(a, x):
    return a ** x 

x_data = list(range(101))
x_data = [x/10 for x in x_data]
y_data = [func1(2, x) for x in x_data]

plt.plot(x_data, y_data, label = 'a = 2')

y_data1 = [func1(2.3, x) for x in x_data]
plt.plot(x_data, y_data1, label = 'a = 2.3')
plt.legend()
```

```python
def e_func(m):
    return (1 + 1 / m) ** m

x_data = list(range(1,1001))
x_data = [x / 10 for x in x_data]
y_data = [e_func(x) for x in x_data]

plt.plot(x_data, y_data)
```

```python
import math
def gauss_func(x, mean, std):
    coeff = 1 / (std * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * (std ** 2))
    
    return coeff * math.exp(exponent)
mean = 10
std = 4
x_data = list(range(1, 200))
x_data = [x / 10 for x in x_data]
y_data = [gauss_func(x, mean, std) for x in x_data]

plt.plot(x_data, y_data)
```

Hyperbolic Functions

```python
'''
Defined in terms of certain combinations of e**x and e**(-x)
+ Hyperbolic cosine
+ Hyperbolic sine
+ Hyperbolic tangent
'''

# Softmax function: chuyn cc gi tr ca mt vector thnh cc gi tr xc sut
def naive_softmax(data):
    exp_data = [math.exp(x) for x in data]
    sum_data = sum(exp_data)
    result = [x / sum_data for x in exp_data]
    return result

data = [1, 2, 3]
print(naive_softmax(data))

# Softmax function (stable)
def stable_softmax(data):
    max_value = max(data)
    exp_data = [math.exp(x - max_value) for x in data]
    sum_data = sum(exp_data)
    result = [x / sum_data for x in exp_data]
    return result

data = [1000, 2000, 3000]
print(stable_softmax(data))
```
