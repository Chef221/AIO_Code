Exercise 1: Viết function tìm thể tích

```python
import math
# Hình lập phương
def calculate_volume_1(s):
    return s ** 3
# Hình hộp chữ nhật
def calculate_volume_2(l, w, h):
    return l * w * h
# Hình trụ tròn
def calculate_volume_3(r, h):
    return math.pi * (r ** 2) * h
# Hình cầu
def calculate_volume_4(r):
    return 4/3 * math.pi * (r ** 3)

print('Diện tích hình lập phương: {}'.format(calculate_volume_1(2)))
print('Diện tích hình hộp chữ nhật: {}'.format(calculate_volume_2(2, 3, 4)))
print('Diện tích hình trụ tròn: {}'.format(round(calculate_volume_3(2, 3), 2)))
print('Diện tích hình cầu: {}'.format(round(calculate_volume_4(2),2)))
```

Exercise 2: Viết function đánh giá classification model bằng f1-score


```python
def calculate_f1_score(tp, fp, fn):
    if type(tp) != int:
        print('tp must be int')
        return None
    elif type(fp) != int:
        print('fp must be int')
        return None
    elif type(fn) != int:
        print('fn must be int')
        return None
    elif tp <= 0 or fp <= 0 or fn <= 0:
        print('tp, fp, fn must be greater than 0')
        return None
    else:
        precision = tp / (tp +  fp)
        recall = tp / (tp + fn)
        f1_score = (2 * precision * recall) / (precision + recall) 

        print(f'Precision: {precision}')
        print(f'Recall: {recall}')
        print(f'F1 Score: {f1_score}') 

calculate_f1_score(2, 3, 4) 
```

Exercise 3: Viết function tính kết quả theo các activation function

```python
import math
def is_number(n):
    try:
        float(n)
    
    except ValueError:
        return False
    return True

def calc_activation_func():
    x = input('Nhập x: ')
    if is_number(x) == False:
        print('x must be a number')
        return None
    else:
        x = float(x) 
        print('Input x =', x)
    
    func = input('Input activation Function (binary|sigmoid|elu):')
    if func != 'binary' and func != 'sigmoid' and func != 'elu':
        print(f'{func} is not supported')
        return None
    else: print(f'Activation function: {func}')
    result = None
    if func == 'binary':
        if x < 0: 
            result = 0
            print('binary: f({}) = {}'.format(x, result))
        else: 
            result = 1
            print('binary: f({}) = {}'.format(x, result))
    elif func == 'sigmoid':
        result = 1 / (1 + math.exp(-x))
        print('sigmoid: f({}) = {}'.format(x, result))
    else:
        if x < 0:
            result = 0.01 * (math.e ** x - 1)
            print('elu: f({}) = {}'.format(x, result))
        else: 
            result = x
            print('elu: f({}) = {}'.format(x, result))

calc_activation_func()
```

Exercise 4: Viết function tính loss theo regression loss function


```python
import random
import math

def calc_MAE_Loss(num_samples, targets, predicts):
    result = 0
    for idx in range(num_samples):
        result += abs(targets[idx] - predicts[idx])
    return result / num_samples
def calc_MSE_Loss(num_samples, targets, predicts):
    result = 0
    for idx in range(num_samples):
        result += (targets[idx] - predicts[idx]) ** 2
    return result / num_samples
def calc_RMSE_Loss(num_samples, targets, predicts):
    result = calc_MSE_Loss(num_samples=num_samples, targets=targets, predicts=predicts)
    return math.sqrt(result)

def Exercise4():
    num_samples = input('Input number of samples: ')
    if num_samples.isnumeric() == False:
        print('Number of samples must be an integer')
        return None
    else:   
        num_samples = int(num_samples)
        print(f'Number of samples: {num_samples}')
    loss_name = input("Input loss name (MAE | MSE | RMSE): ")
    if loss_name != 'MAE' and loss_name != 'MSE' and loss_name != 'RMSE':
        print(f'{loss_name} is not supported')
        return None
    else:
        print(f'Loss function: {loss_name}')
    
    predicts = [None] * num_samples
    targets = [None] * num_samples
    for idx in range(num_samples):
        predicts[idx] = random.uniform(0, 10) 
        targets[idx] = random.uniform(0, 10) 
    
    print('Predict:',predicts)
    print('Target:', targets)
    
    result = None
    if loss_name == 'MAE': 
        result = calc_MAE_Loss(num_samples, targets, predicts)
        print(f'MAE Loss: {result}')
    elif loss_name == 'MSE':
        result = calc_MSE_Loss(num_samples, targets, predicts)
        print(f'MSE Loss: {result}')
    else:
        result = calc_RMSE_Loss(num_samples, targets, predicts)
        print(f'RMSE Loss: {result}')
    
Exercise4()
```

Exercise 6: Viết function đảo ngược các vị trí 1 số nguyên dương cho trước

```python
def reverse_num(x):
    result = ""
    while x != 0:
        result += str(x % 10)
        x //= 10
    return int(result)
print(reverse_num(6540))
```

```python
# Viết function trả về một dictionary đếm số lượng chữ xuất hiện trong một từ, với key
# chữ cái và value là số lần xuất hiện
def count_freq(word):
    word = word.lower()
    a_dict = {}
    for char in word:
        a_dict[char] = a_dict.get(char, 0) + 1
    return a_dict

s = 'Happiness'
print(count_freq(s))

# Viết function trả về một dict đếm số lượng từ xuất hiện trong một câu
def count_freq_para(paragraph):
    paragraph = paragraph.replace('.','').replace('.','').replace('-',' ')
    paragraph = paragraph.lower()
    a_list = paragraph.split()
    a_dict = {}
    for word in a_list:
        a_dict[word] = a_dict.get(word, 0) + 1
    return a_dict
```

```python
'''
Getting Max over Kernel
Problem 01: Cho một list các số nguyên num_list và một sliding window (các bạn có thể hiểu
sliding window như là một list có kích thước nhỏ hơn num_list) có kích thước size k di 
chuyển từ trái sang phải. Mỗi lần dịch chuyển 1 vị trí sang phải có thể nhìn thấy được k
số trong num_list và tìm số lớn nhất trong k số này sau mỗi lần trượt, k phải lớn hơn 
hoặc bằng 1. Các bạn hãy viết chương trình Python giải quyết vấn đề trên

Example: 
- Input: 
    + num_list = [3, 4, 5, 1, -44, 5, 10, 12, 33, 1]
    + k = 3
- Output:
    + [5, 5, 5, 5, 10, 12, 33, 33]
'''
def GettingMaxOverKernel(num_list, k):
    result = []
    for i in range(len(num_list)):
        if i + k > len(num_list):
            break
        else:
            result.append(max(num_list[i:i + k]))
    return result

num_list = [3, 4, 5, 1, -44, 5, 10, 12, 33, 1]
k = 3
print(GettingMaxOverKernel(num_list, k))
```
